### 4.2 Visulation graphique de la distribution de certaines données en fonction des précédentes observations liées à l'analyse des statistiques descriptives

```python
# Analyse de la distribution de la variable surface de la géométrie des groupes de bâtiments

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(bat_gpe["s_geom_groupe"], kde=False)
plt.title(f"Histogramme de {"s_geom_groupe"}")
    
plt.subplot(1,2,2)
sns.boxplot(x=bat_gpe["s_geom_groupe"])
plt.title(f"Boxplot de {"s_geom_groupe"}")
plt.show()
```

**Interprétation des résultats**

La visualisation de la distribution montre bien une concentration des surfaces des groupes de bâtiments vers des valeurs faibles (qu'on pourrait qualifier de normales car étant attendues) et la présence de valeurs extrêmes qui pourraient correspondre à des ensembles immobiliers particuliers (centre commercial, plateforme logistique, etc.).

Cette hypothèse pourra être vérifiée en recoupant le ou les eregistrements correspondants à ces valeurs aberrantes avec les informations contenues dans d'autres tables, comme celles relatives aux adresses et aux usages des groupes de bâtiments par exemple.

```python
# Analyse de la distribution des variables de valeurs foncières et de prix au m²

for var in bat_gpe_dvf_stat:
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    sns.histplot(bat_gpe_dvf_stat[var], kde=False)
    plt.title(f"Histogramme de {var}")
    
    plt.subplot(1,2,2)
    sns.boxplot(x=bat_gpe_dvf_stat[var])
    plt.title(f"Boxplot de {var}")
    plt.show()
```

**Interprétation des résultats**

Ici, les visualisations réalisées sur les variables batiment_groupe_id et code_departement_insee n'ont aucune portée scientifique.
Aussi, pour rappel, les données relatives aux nombres de mutations ne présentent pas d'intérêt particulier pour l'étude. 

Les visualisations de la distribution pour les valeurs foncières et les prix au m² montrent bien une concentration vers des valeurs faibles (qu'on pourrait qualifier de normales car attendues). Cependant, il est également relevé la présence d'une ou plusieurs valeurs extrêmes selon les variables. Et une valeur particulièrement aberrante est relevée pour chacune des variables de valeur foncière et de prix du local au m². 

Il conviendra donc de recouper les enregistrements correspondant à ces valeurs aberrantes avec les informations contenues dans d'autres tables, comme celles relatives aux adresses et aux usages des groupes de bâtiments par exemple, pour comprendre leur nature.
